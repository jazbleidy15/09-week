Manual: Gestión de Fechas y Tiempos en Python
Introducción
En este manual exploraremos el uso de funciones de tiempo y jerarquía de fechas utilizando Python y pandas. Este conocimiento es fundamental en ciencia de datos, ya que muchas veces se trabaja con datos temporales, como series de tiempo, registros de sensores o ventas a lo largo del tiempo.
1. Funciones de Tiempo

Descripción
Las funciones de tiempo permiten menipular y extraer información de fecas y horas. Con pandas, podemos generar fechas, transformarlas y obtener componentes como el año, mes, día, o nombre del día de la semana.

Aplicabilidad


*   Análisis financiero
*   Estudio de tendencias


*   Segumiento de sensores
*   Planificación y pronóstico





```python
import pandas as pd
import pandas as pd

# Crear un rango de fechas
fechas = pd.date_range(start='2023-01-01',periods=10, freq='D')
print(fechas)

# Extraer elementos
df = pd.DataFrame({'fechas':fechas})
df['año']=df['fechas'].dt.year
df['mes']=df['fechas'].dt.month
df['dia']=df['fechas'].dt.day
df['día_semana']=df['fechas'].dt.day_name()
df
df = pd.DataFrame({'fechas':fechas})
df['año']=df['fechas'].dt.year
df['mes']=df['fechas'].dt.month
df['dia']=df['fechas'].dt.day
df['día_semana']=df['fechas'].dt.day_name()
df
```

Jerarquía de Fechas

Descripción

La jerarquía de fechas permite agrupar datos según niveles temporales como año, trimestre, mes o semana. Esto es clave para organizar la información y analizarla por periodos.

Aplicabilidad



*   Dashboards
*   Reportes mensuales o trimestrales

*   Análisis de ventas o consumo energético

Actividad: Agrupación temporal y visualización



*   Agrupa un dataset por trimestre y calcula estadística descriptivas
*   Visualiza las ventas promedio por mes



```python
# Agrupar por trimestre y describir
df['trimestre'] = df['fecha'].dt.to_period('Q')
# Changed 'ventas' to 'Ventas' to match the actual column name in the DataFrame.
estadísticas_trimestrales = df.groupby('trimestre')['Ventas'].describe()
estadísticas_trimestrales
df.groupby('trimestre')['Ventas'].describe() # Changed 'ventas' to 'Ventas' here as well.
estadísticas_trimestrales
```

```python
import matplotlib.pyplot as plt

# Visualización promedio mensual
# Changed 'ventas' to 'Ventas' to match the actual column name in the DataFrame.
ventas_promedio = df.groupby(df['fecha'].dt.to_period('M'))['Ventas'].mean()
ventas_promedio.plot(kind='bar',
title='ventas Promedio por Mes',
ylabel='ventas promedio', xlabel='Mes')
plt.tight_layout()
plt.show()
```
